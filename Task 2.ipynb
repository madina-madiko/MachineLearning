```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate, linalg

X = np.arange(1, 16, 1)
Y = np.sin(X/5)*np.exp(X/10)+5*np.exp(-X/2)
A = np.array([[3,1], [3,15]])
b = np.array([float(Y[0]),float(Y[14])])
b = b.reshape((2,1))
print(A)
res = linalg.solve(A, b)
plt.plot(X, Y)
plt.plot([1,15], b)
plt.show()
print(res)
```

```python
A = np.array([[1,1,1],[1,8,8**2],[1,15,15**2]])
b = np.array([float(Y[0]),float(Y[1]), float(Y[14])])
b = b.reshape(3,1)
print(A)
res = linalg.solve(A, b)
plt.plot(X, Y)
plt.plot([1,8,15], b)
plt.show()
print(res)
```

```python
A = np.array([[1,1,1,1],[1,4,4**2,4**3],[1,10,10**2,10**3],[1,15,15**2,15**3]])
b = np.array([float(Y[0]),float(Y[1]),float(Y[7]), float(Y[14])])
b = b.reshape(4,1)
print(A)
res = linalg.solve(A, b)
plt.plot(X, Y)
plt.plot([1,4,10,15], b)
plt.show()
print("Results: \n", "w_0: ", res[0], "\n",  "w_1: ", res[1], "\n", "w_2: ", res[2], "\n", "w_3: ", res[3])
```
